```python
import os
import shutil

# 当前工作目录
current_dir = os.getcwd()
print(current_dir)
```

```python
# 定义需要清空的主文件夹名称
folders_to_clear = ["../collectData", "../collectData2"]

# 子文件夹的名称列表
subfolders = ["BaseStationPosition", "UserEquipmentPosition", "DataRate", "UserQoE"]

# 遍历每个主文件夹
for folder in folders_to_clear:
    folder_path = os.path.join(current_dir, folder)  # 拼接完整路径

    # 检查主文件夹是否存在
    if not os.path.exists(folder_path):
        print(f"文件夹不存在：{folder_path}")
        continue

    # 遍历子文件夹
    for subfolder in subfolders:
        subfolder_path = os.path.join(folder_path, subfolder)

        # 检查子文件夹是否存在
        if not os.path.exists(subfolder_path):
            print(f"子文件夹不存在：{subfolder_path}")
            continue

        # 清空子文件夹内容
        for item in os.listdir(subfolder_path):
            item_path = os.path.join(subfolder_path, item)
            if os.path.isdir(item_path):
                shutil.rmtree(item_path)  # 删除文件夹
                # print(f"已删除文件夹：{item_path}")
            elif os.path.isfile(item_path):
                os.remove(item_path)  # 删除文件
                # print(f"已删除文件：{item_path}")

        print(f"已清空子文件夹：{subfolder_path}")
```

```python
from mobile_env.scenarios.custom import MComCustom

# 创建环境实例
env = MComCustom(render_mode="rgb_array")
```

```python
# 一共有多少轮迭代
iteration_number = 1000
# 每次迭代用户要运动多少步
step_number = 20
env.reset()
```

```python
import matplotlib.pyplot as plt
from IPython import display

for epoch_number in range(iteration_number):
    env.reset()
    env.save_base_station_positions(epoch_number)

    for curr_step in range(step_number):
        env.step(epoch_number, curr_step)
        # plt.imshow(env.render())
        # display.display(plt.gcf())
        # display.clear_output(wait=False)
    
    env.save_epoch_data(epoch_number)
    print("第", epoch_number, "轮数据收集完成")

# print("数据收集完成。")
```
